Exploratory Data Analysis Using Superstore sales data :

## Business Objective

The client is a nationwide retail company that sells a wide variety of office supplies, furniture, and technology products through both online and offline channels. As part of its ongoing strategy to enhance performance, customer satisfaction, and profitability, the company has requested a detailed exploratory data analysis (EDA) of its sales data.

The primary goal of this analysis is to uncover actionable insights that can support data-driven decision-making. Specifically, the client wants to:

1. **Evaluate Sales Performance**  
   Understand how different product categories, customer segments, and regions are performing in terms of revenue.

2. **Analyze Trends and Seasonality**  
   Identify temporal patterns in sales over months and years to guide promotional campaigns and inventory planning.

3. **Assess Operational Efficiency**  
   Review shipping times and logistics by examining shipping durations, methods, and delays.

4. **Explore Customer Behavior**  
   Investigate purchasing trends among customers, including high-value customers and frequent buyers.

5. **Uncover Improvement Areas**  
   Pinpoint underperforming products, locations, or segments that may require strategic focus or intervention.

In this project, I will  visualize and present data in a clear and interactive manner to generate insights which can support strategic decisions across sales, marketing, and operations.

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'])
df['Postal_Code'] = df['Postal_Code'].astype("object")
```

```python
df.dtypes
```

```python
##Step 3: Feature Engineering
```

```python
# Create new features
df['Shipping_Duration'] = (df['Ship_Date'] - df['Order_Date']).dt.days
df['Order_Year'] = df['Order_Date'].dt.year
df['Order_Month'] = df['Order_Date'].dt.month_name()

df[['Order_Date', 'Ship_Date', 'Shipping_Duration', 'Order_Year', 'Order_Month']].head()
```

# Exploratory Data Analysis (EDA)

```python
category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
category
```

```python
#df.groupby('Category').agg({"Sales":"sum"})
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
```

```python
shipping_duration_by_mode = df.groupby('Ship_Mode')['Shipping_Duration'].mean().sort_values()
```

```python
top_subcategories = df.groupby('Sub_Category')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
sns.barplot(x=category.values, y=category.index, palette="bright")
```

```python
#Sales by Region
sns.barplot(x=region_sales.values, y=region_sales.index, palette="mako")
```

```python
# 3. Sales by Segment
sns.barplot(x=segment_sales.values, y=segment_sales.index, palette="rocket")
```

```python
# 4. Avg. Shipping Duration by Ship Mode
sns.barplot(x=shipping_duration_by_mode.values, y=shipping_duration_by_mode.index,  palette="coolwarm")
```

```python
shipping_duration_by_mode
```

```python
# 5. Top 10 Sub-Categories by Sales
sns.barplot(x=top_subcategories.values, y=top_subcategories.index, palette="cubehelix")
```

```python
# 6. Sales Distribution Over Time
monthly_sales_trend = df.set_index('Order_Date').resample('M')['Sales'].sum().reset_index()
```

```python
monthly_sales_trend
```

```python
sns.barplot(x=monthly_sales_trend["Sales"],y = monthly_sales_trend['Order_Date'],palette="deep")
```

```python
monthly_sales_trend['Month_Year'] = monthly_sales_trend['Order_Date'].dt.strftime('%b %Y')
```

```python
sns.barplot(x='Month_Year', y='Sales', data=monthly_sales_trend, palette="deep")

```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(x='Month_Year', y='Sales', data=monthly_sales_trend, palette="deep")

plt.title("Monthly Sales Trend", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
